# Oxford Pet Binary Segmentation

This notebook implements a PyTorch pipeline for binary semantic segmentation of pet foreground vs background on the Oxford-IIIT Pet dataset.

Implemented models:
- FCN-ResNet18
- SegNet-VGG16
- HRNet-W18

Experimental controls:
- fixed train/validation/test split from preprocessing_artifacts`n- image size 320x320`n- trimap boundary pixels treated as foreground
- prediction threshold  .5`n- final comparison ranked by test mIoU

### 3. Dataset and Shared Split Resolution


```python
def has_oxford_pet_layout(path: Path) -> bool:
    return (
        (path / "images").is_dir()
        and (path / "annotations" / "trimaps").is_dir()
        and (path / "annotations" / "trainval.txt").is_file()
        and (path / "annotations" / "test.txt").is_file()
    )


def resolve_dataset_root(project_root: Path) -> Path:
    candidates = [
        project_root / "data" / "oxford_pet" / "oxford_pet",
        project_root / "data" / "oxford_pet",
        project_root / "oxford_pet" / "oxford_pet",
        project_root / "oxford_pet",
    ]
    checked = []
    for candidate in candidates:
        candidate = candidate.resolve()
        nested = (candidate / "oxford_pet").resolve()
        for path in (candidate, nested):
            if path in checked:
                continue
            checked.append(path)
            if has_oxford_pet_layout(path):
                return path
    raise FileNotFoundError(f"Oxford-IIIT Pet dataset not found. Checked: {checked}")


def resolve_split_dir(project_root: Path) -> Path:
    split_dir = project_root / "preprocessing_artifacts"
    required = {"train_split.csv", "val_split.csv", "test_split.csv"}
    if not split_dir.is_dir():
        raise FileNotFoundError(f"Missing split directory: {split_dir}")
    missing = [name for name in required if not (split_dir / name).is_file()]
    if missing:
        raise FileNotFoundError(f"Missing split CSVs in {split_dir}: {missing}")
    return split_dir


def seed_everything(seed: int) -> None:
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)


seed_everything(config.seed)
DATASET_ROOT = resolve_dataset_root(config.project_root)
SPLIT_DIR = resolve_split_dir(config.project_root)
IMAGE_DIR = DATASET_ROOT / "images"
TRIMAP_DIR = DATASET_ROOT / "annotations" / "trimaps"

print("Dataset root:", DATASET_ROOT)
print("Split dir:", SPLIT_DIR)
```

### 5. PyTorch Dataset and Dataloaders


```python
def pil_to_normalized_tensor(image: Image.Image) -> torch.Tensor:
    image_np = np.asarray(image, dtype=np.float32) / 255.0
    image_tensor = torch.from_numpy(image_np).permute(2, 0, 1).contiguous()
    return (image_tensor - IMAGENET_MEAN) / IMAGENET_STD


def trimap_to_binary_tensor(trimap: Image.Image, boundary_mode: BoundaryMode) -> torch.Tensor:
    trimap_np = np.asarray(trimap, dtype=np.uint8)
    if boundary_mode == "foreground":
        mask = np.where((trimap_np == 1) | (trimap_np == 3), 1.0, 0.0).astype(np.float32)
        return torch.from_numpy(mask).unsqueeze(0)
    if boundary_mode == "ignore":
        mask = np.where(trimap_np == 3, IGNORE_INDEX, np.where(trimap_np == 1, 1, 0)).astype(np.int64)
        return torch.from_numpy(mask).unsqueeze(0)
    raise ValueError(f"Unsupported boundary_mode: {boundary_mode}")


class OxfordPetBinaryDataset(Dataset):
    def __init__(
        self,
        split_df: pd.DataFrame,
        dataset_root: Path,
        image_size: tuple[int, int],
        boundary_mode: BoundaryMode,
        training: bool,
    ) -> None:
        self.df = split_df.reset_index(drop=True).copy()
        self.dataset_root = dataset_root
        self.image_size = image_size
        self.boundary_mode = boundary_mode
        self.training = training
        self.image_dir = self.dataset_root / "images"
        self.trimap_dir = self.dataset_root / "annotations" / "trimaps"

    def __len__(self) -> int:
        return len(self.df)

    def _load_pair(self, image_id: str) -> tuple[Image.Image, Image.Image]:
        image = Image.open(self.image_dir / f"{image_id}.jpg").convert("RGB")
        trimap = Image.open(self.trimap_dir / f"{image_id}.png")
        return image, trimap

    def _apply_transforms(self, image: Image.Image, trimap: Image.Image) -> tuple[Image.Image, Image.Image]:
        target_size = (self.image_size[1], self.image_size[0])

        if self.training and random.random() < 0.5:
            image = image.transpose(Image.FLIP_LEFT_RIGHT)
            trimap = trimap.transpose(Image.FLIP_LEFT_RIGHT)

        if self.training:
            angle = random.uniform(-10.0, 10.0)
            image = image.rotate(angle, resample=Image.BILINEAR)
            trimap = trimap.rotate(angle, resample=Image.NEAREST, fillcolor=2)

            if random.random() < 0.5:
                image = ImageEnhance.Brightness(image).enhance(random.uniform(0.9, 1.1))
            if random.random() < 0.5:
                image = ImageEnhance.Contrast(image).enhance(random.uniform(0.9, 1.1))

        image = image.resize(target_size, resample=Image.BILINEAR)
        trimap = trimap.resize(target_size, resample=Image.NEAREST)
        return image, trimap

    def __getitem__(self, index: int) -> tuple[torch.Tensor, torch.Tensor]:
        row = self.df.iloc[index]
        image, trimap = self._load_pair(row["image_id"])
        image, trimap = self._apply_transforms(image, trimap)
        image_tensor = pil_to_normalized_tensor(image)
        mask_tensor = trimap_to_binary_tensor(trimap, self.boundary_mode)
        return image_tensor, mask_tensor


train_dataset = OxfordPetBinaryDataset(train_df, DATASET_ROOT, config.image_size, config.boundary_mode, training=True)
val_dataset = OxfordPetBinaryDataset(val_df, DATASET_ROOT, config.image_size, config.boundary_mode, training=False)
test_dataset = OxfordPetBinaryDataset(test_df, DATASET_ROOT, config.image_size, config.boundary_mode, training=False)

train_loader = DataLoader(train_dataset, batch_size=config.batch_size, shuffle=True, num_workers=config.num_workers)
val_loader = DataLoader(val_dataset, batch_size=config.eval_batch_size, shuffle=False, num_workers=config.num_workers)
test_loader = DataLoader(test_dataset, batch_size=config.eval_batch_size, shuffle=False, num_workers=config.num_workers)

images, masks = next(iter(train_loader))
print("Train samples:", len(train_dataset))
print("Val samples:", len(val_dataset))
print("Test samples:", len(test_dataset))
print("Images:", images.shape, images.dtype)
print("Masks:", masks.shape, masks.dtype)
print("Mask values:", torch.unique(masks)[:10])
```

## 6. FCN-ResNet18 Model


```python
class FCNResNet18(nn.Module):
    def __init__(self, pretrained_backbone: bool = True) -> None:
        super().__init__()
        weights = ResNet18_Weights.DEFAULT if pretrained_backbone else None
        backbone = resnet18(weights=weights)

        self.stem = nn.Sequential(
            backbone.conv1,
            backbone.bn1,
            backbone.relu,
            backbone.maxpool,
        )
        self.layer1 = backbone.layer1
        self.layer2 = backbone.layer2
        self.layer3 = backbone.layer3
        self.layer4 = backbone.layer4

        self.proj2 = nn.Sequential(nn.Conv2d(128, 128, kernel_size=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True))
        self.proj3 = nn.Sequential(nn.Conv2d(256, 128, kernel_size=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True))
        self.proj4 = nn.Sequential(nn.Conv2d(512, 128, kernel_size=1), nn.BatchNorm2d(128), nn.ReLU(inplace=True))

        self.fuse3 = nn.Sequential(
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
        )
        self.fuse2 = nn.Sequential(
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
        )
        self.classifier = nn.Conv2d(64, 1, kernel_size=1)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        input_size = x.shape[-2:]

        x = self.stem(x)
        x = self.layer1(x)
        c2 = self.layer2(x)
        c3 = self.layer3(c2)
        c4 = self.layer4(c3)

        y = self.proj4(c4)
        y = F.interpolate(y, size=c3.shape[-2:], mode="bilinear", align_corners=False)
        y = self.fuse3(y + self.proj3(c3))

        y = F.interpolate(y, size=c2.shape[-2:], mode="bilinear", align_corners=False)
        y = self.fuse2(y + self.proj2(c2))

        logits = self.classifier(y)
        logits = F.interpolate(logits, size=input_size, mode="bilinear", align_corners=False)
        return logits


class DiceLoss(nn.Module):
    def __init__(self, smooth: float = 1.0) -> None:
        super().__init__()
        self.smooth = smooth

    def forward(self, logits: torch.Tensor, targets: torch.Tensor) -> torch.Tensor:
        probs = torch.sigmoid(logits)
        probs = probs.reshape(probs.shape[0], -1)
        targets = targets.reshape(targets.shape[0], -1)

        intersection = (probs * targets).sum(dim=1)
        denominator = probs.sum(dim=1) + targets.sum(dim=1)
        dice = (2.0 * intersection + self.smooth) / (denominator + self.smooth)
        return 1.0 - dice.mean()
```

### 7. Metrics and Evaluation Helpers


```python
def confusion_from_logits(logits: torch.Tensor, targets: torch.Tensor, threshold: float) -> dict[str, float]:
    probs = torch.sigmoid(logits)
    preds = (probs >= threshold).float()
    targets = targets.float()

    tp = ((preds == 1) & (targets == 1)).sum().item()
    tn = ((preds == 0) & (targets == 0)).sum().item()
    fp = ((preds == 1) & (targets == 0)).sum().item()
    fn = ((preds == 0) & (targets == 1)).sum().item()
    return {"tp": tp, "tn": tn, "fp": fp, "fn": fn}


def metrics_from_confusion(tp: float, tn: float, fp: float, fn: float) -> dict[str, float]:
    eps = 1e-8
    pet_iou = tp / (tp + fp + fn + eps)
    bg_iou = tn / (tn + fp + fn + eps)
    miou = 0.5 * (pet_iou + bg_iou)
    dice = (2 * tp) / (2 * tp + fp + fn + eps)
    pixel_acc = (tp + tn) / (tp + tn + fp + fn + eps)
    precision = tp / (tp + fp + eps)
    recall = tp / (tp + fn + eps)
    return {
        "mIoU": miou,
        "Pet IoU": pet_iou,
        "Background IoU": bg_iou,
        "Dice/F1": dice,
        "Pixel Accuracy": pixel_acc,
        "Foreground Precision": precision,
        "Foreground Recall": recall,
    }


def evaluate_loader(
    model: nn.Module,
    loader: DataLoader,
    bce_loss_fn: nn.Module,
    dice_loss_fn: nn.Module,
    device: torch.device,
    threshold: float,
) -> dict[str, float]:
    model.eval()
    total_loss = 0.0
    total_tp = total_tn = total_fp = total_fn = 0.0

    with torch.no_grad():
        for images, masks in loader:
            images = images.to(device)
            masks = masks.to(device)
            logits = model(images)

            loss = bce_loss_fn(logits, masks) + dice_loss_fn(logits, masks)
            total_loss += loss.item() * images.size(0)

            conf = confusion_from_logits(logits, masks, threshold)
            total_tp += conf["tp"]
            total_tn += conf["tn"]
            total_fp += conf["fp"]
            total_fn += conf["fn"]

    metrics = metrics_from_confusion(total_tp, total_tn, total_fp, total_fn)
    metrics["loss"] = total_loss / len(loader.dataset)
    return metrics
```

### 8. Training Helpers


```python
CHECKPOINT_DIR = PROJECT_ROOT / "checkpoints"
CHECKPOINT_DIR.mkdir(parents=True, exist_ok=True)
BEST_MODEL_PATH = CHECKPOINT_DIR / "fcn_resnet18_best.pt"


def build_optimizer(model: FCNResNet18, config: Config) -> torch.optim.Optimizer:
    backbone_params = []
    head_params = []

    for name, param in model.named_parameters():
        if name.startswith(("stem", "layer1", "layer2", "layer3", "layer4")):
            backbone_params.append(param)
        else:
            head_params.append(param)

    return torch.optim.AdamW(
        [
            {"params": backbone_params, "lr": config.backbone_lr},
            {"params": head_params, "lr": config.head_lr},
        ],
        weight_decay=config.weight_decay,
    )


def train_one_epoch(
    model: nn.Module,
    loader: DataLoader,
    optimizer: torch.optim.Optimizer,
    bce_loss_fn: nn.Module,
    dice_loss_fn: nn.Module,
    device: torch.device,
) -> tuple[float, float]:
    model.train()
    total_loss = 0.0
    start = perf_counter()

    for images, masks in loader:
        images = images.to(device)
        masks = masks.to(device)

        optimizer.zero_grad(set_to_none=True)
        logits = model(images)
        loss = bce_loss_fn(logits, masks) + dice_loss_fn(logits, masks)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * images.size(0)

    epoch_time = perf_counter() - start
    avg_loss = total_loss / len(loader.dataset)
    return avg_loss, epoch_time


def fit_model(
    model: FCNResNet18,
    train_loader: DataLoader,
    val_loader: DataLoader,
    config: Config,
    device: torch.device,
) -> pd.DataFrame:
    bce_loss_fn = nn.BCEWithLogitsLoss()
    dice_loss_fn = DiceLoss()
    optimizer = build_optimizer(model, config)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="max", factor=0.5, patience=3)

    best_val_miou = -1.0
    best_state = None
    patience_counter = 0
    history = []

    for epoch in range(1, config.max_epochs + 1):
        train_loss, epoch_time = train_one_epoch(model, train_loader, optimizer, bce_loss_fn, dice_loss_fn, device)
        val_metrics = evaluate_loader(model, val_loader, bce_loss_fn, dice_loss_fn, device, config.threshold)
        scheduler.step(val_metrics["mIoU"])

        current_lrs = [group["lr"] for group in optimizer.param_groups]
        row = {
            "epoch": epoch,
            "train_loss": train_loss,
            "val_loss": val_metrics["loss"],
            "val_mIoU": val_metrics["mIoU"],
            "val_Pet IoU": val_metrics["Pet IoU"],
            "val_Dice/F1": val_metrics["Dice/F1"],
            "epoch_time_sec": epoch_time,
            "backbone_lr": current_lrs[0],
            "head_lr": current_lrs[1],
        }
        history.append(row)
        display(pd.DataFrame([row]))

        if val_metrics["mIoU"] > best_val_miou:
            best_val_miou = val_metrics["mIoU"]
            best_state = copy.deepcopy(model.state_dict())
            torch.save(
                {
                    "epoch": epoch,
                    "model_state": best_state,
                    "best_val_miou": best_val_miou,
                    "config": config.__dict__,
                },
                BEST_MODEL_PATH,
            )
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= config.early_stopping_patience:
            print(f"Early stopping at epoch {epoch}.")
            break

    if best_state is not None:
        model.load_state_dict(best_state)

    return pd.DataFrame(history)


def make_tiny_overfit_loader(dataset: Dataset, n_samples: int = 8) -> DataLoader:
    subset_indices = list(range(min(n_samples, len(dataset))))
    subset = Subset(dataset, subset_indices)
    return DataLoader(subset, batch_size=min(config.batch_size, len(subset)), shuffle=True, num_workers=0)
```

### 10. Final Test Evaluation and Efficiency Metrics


```python
def measure_inference_time(model: nn.Module, loader: DataLoader, device: torch.device, warmup: int = 10, timed: int = 100) -> float:
    model.eval()
    timings = []
    seen = 0

    with torch.no_grad():
        for idx, (images, _) in enumerate(loader):
            images = images.to(device)
            start = perf_counter()
            _ = model(images)
            elapsed = perf_counter() - start

            if idx >= warmup:
                timings.append(elapsed / images.size(0))
                seen += images.size(0)
            if seen >= timed:
                break

    return float(np.mean(timings)) if timings else float('nan')


checkpoint = torch.load(BEST_MODEL_PATH, map_location=DEVICE, weights_only=False)
model = FCNResNet18(pretrained_backbone=config.use_pretrained_backbone).to(DEVICE)
model.load_state_dict(checkpoint['model_state'])

bce_loss_fn = nn.BCEWithLogitsLoss()
dice_loss_fn = DiceLoss()
test_metrics = evaluate_loader(model, test_loader, bce_loss_fn, dice_loss_fn, DEVICE, config.threshold)

avg_epoch_time = float(history_df['epoch_time_sec'].mean()) if 'history_df' in globals() else float('nan')
inference_time = measure_inference_time(model, test_loader, DEVICE)

results_df = pd.DataFrame(
    [
        {
            'Model': 'FCN-ResNet18',
            'Backbone Init': 'pretrained',
            'Test mIoU': test_metrics['mIoU'],
            'Pet IoU': test_metrics['Pet IoU'],
            'Dice/F1': test_metrics['Dice/F1'],
            'Pixel Accuracy': test_metrics['Pixel Accuracy'],
            'Foreground Precision': test_metrics['Foreground Precision'],
            'Foreground Recall': test_metrics['Foreground Recall'],
            'Parameter Count': total_params,
            'Inference Time / Image (sec)': inference_time,
            'Training Time / Epoch (sec)': avg_epoch_time,
        }
    ]
)
display(results_df)
```

### 11. Qualitative Predictions


```python
def collect_predictions(model: nn.Module, loader: DataLoader, device: torch.device, n_items: int = 6) -> list[dict[str, np.ndarray]]:
    model.eval()
    collected = []
    with torch.no_grad():
        for images, masks in loader:
            images = images.to(device)
            logits = model(images)
            probs = torch.sigmoid(logits)
            preds = (probs >= config.threshold).float()

            for i in range(images.size(0)):
                collected.append(
                    {
                        'image': denormalize_image(images[i].cpu()),
                        'mask': masks[i].squeeze().cpu().numpy(),
                        'pred': preds[i].squeeze().cpu().numpy(),
                    }
                )
                if len(collected) >= n_items:
                    return collected
    return collected


def show_predictions(items: list[dict[str, np.ndarray]]) -> None:
    if not items:
        print('No prediction items to show.')
        return

    fig, axes = plt.subplots(len(items), 4, figsize=(16, 4 * len(items)))
    if len(items) == 1:
        axes = np.expand_dims(axes, axis=0)

    for row_idx, item in enumerate(items):
        image = item['image']
        mask = item['mask']
        pred = item['pred']

        axes[row_idx, 0].imshow(image)
        axes[row_idx, 0].set_title('Input Image')
        axes[row_idx, 1].imshow(mask, cmap='gray')
        axes[row_idx, 1].set_title('Ground Truth')
        axes[row_idx, 2].imshow(pred, cmap='gray')
        axes[row_idx, 2].set_title('Prediction')
        axes[row_idx, 3].imshow(image)
        axes[row_idx, 3].imshow(pred == 1, cmap='Reds', alpha=0.35)
        axes[row_idx, 3].set_title('Prediction Overlay')

        for col_idx in range(4):
            axes[row_idx, col_idx].axis('off')

    plt.tight_layout()
    plt.show()


prediction_items = collect_predictions(model, test_loader, DEVICE, n_items=4)
show_predictions(prediction_items)
```

## 12. SegNet-VGG16 Model


```python
from torchvision.models import VGG16_Weights, vgg16


def make_segnet_decoder_block(in_channels: int, out_channels: list[int]) -> nn.Sequential:
    layers = []
    current_channels = in_channels
    for channels in out_channels:
        layers.append(nn.Conv2d(current_channels, channels, kernel_size=3, padding=1))
        layers.append(nn.ReLU(inplace=True))
        current_channels = channels
    return nn.Sequential(*layers)


def init_segnet_decoder(module: nn.Module) -> None:
    if isinstance(module, nn.Conv2d):
        nn.init.kaiming_normal_(module.weight, mode='fan_out', nonlinearity='relu')
        if module.bias is not None:
            nn.init.zeros_(module.bias)


class SegNetVGG16(nn.Module):
    def __init__(self, pretrained_backbone: bool = True) -> None:
        super().__init__()
        weights = VGG16_Weights.DEFAULT if pretrained_backbone else None
        backbone = vgg16(weights=weights)
        features = backbone.features

        self.encoder_block1 = nn.Sequential(*features[0:4])
        self.encoder_block2 = nn.Sequential(*features[5:9])
        self.encoder_block3 = nn.Sequential(*features[10:16])
        self.encoder_block4 = nn.Sequential(*features[17:23])
        self.encoder_block5 = nn.Sequential(*features[24:30])

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, return_indices=True)
        self.unpool = nn.MaxUnpool2d(kernel_size=2, stride=2)

        self.decoder_block5 = make_segnet_decoder_block(512, [512, 512, 512])
        self.decoder_block4 = make_segnet_decoder_block(512, [512, 512, 256])
        self.decoder_block3 = make_segnet_decoder_block(256, [256, 256, 128])
        self.decoder_block2 = make_segnet_decoder_block(128, [128, 64])
        self.decoder_block1 = make_segnet_decoder_block(64, [64, 64])
        self.classifier = nn.Conv2d(64, 1, kernel_size=1)

        self.decoder_block5.apply(init_segnet_decoder)
        self.decoder_block4.apply(init_segnet_decoder)
        self.decoder_block3.apply(init_segnet_decoder)
        self.decoder_block2.apply(init_segnet_decoder)
        self.decoder_block1.apply(init_segnet_decoder)
        self.classifier.apply(init_segnet_decoder)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x1 = self.encoder_block1(x)
        size1 = x1.size()
        x, indices1 = self.pool(x1)

        x2 = self.encoder_block2(x)
        size2 = x2.size()
        x, indices2 = self.pool(x2)

        x3 = self.encoder_block3(x)
        size3 = x3.size()
        x, indices3 = self.pool(x3)

        x4 = self.encoder_block4(x)
        size4 = x4.size()
        x, indices4 = self.pool(x4)

        x5 = self.encoder_block5(x)
        size5 = x5.size()
        x, indices5 = self.pool(x5)

        x = self.unpool(x, indices5, output_size=size5)
        x = self.decoder_block5(x)

        x = self.unpool(x, indices4, output_size=size4)
        x = self.decoder_block4(x)

        x = self.unpool(x, indices3, output_size=size3)
        x = self.decoder_block3(x)

        x = self.unpool(x, indices2, output_size=size2)
        x = self.decoder_block2(x)

        x = self.unpool(x, indices1, output_size=size1)
        x = self.decoder_block1(x)

        return self.classifier(x)
```

### 13. SegNet Training Helpers


```python
SEGNET_BEST_MODEL_PATH = CHECKPOINT_DIR / 'segnet_vgg16_best.pt'


def serialize_config(config: Config) -> dict[str, object]:
    serialized = {}
    for key, value in config.__dict__.items():
        serialized[key] = str(value) if isinstance(value, Path) else value
    return serialized


def build_segnet_optimizer(model: SegNetVGG16, config: Config) -> torch.optim.Optimizer:
    encoder_params = []
    decoder_params = []

    for name, param in model.named_parameters():
        if name.startswith('encoder_block'):
            encoder_params.append(param)
        else:
            decoder_params.append(param)

    return torch.optim.AdamW(
        [
            {'params': encoder_params, 'lr': config.backbone_lr},
            {'params': decoder_params, 'lr': config.head_lr},
        ],
        weight_decay=config.weight_decay,
    )


def fit_segnet_model(
    model: SegNetVGG16,
    train_loader: DataLoader,
    val_loader: DataLoader,
    config: Config,
    device: torch.device,
    checkpoint_path: Path | None = None,
) -> pd.DataFrame:
    checkpoint_path = checkpoint_path or SEGNET_BEST_MODEL_PATH
    bce_loss_fn = nn.BCEWithLogitsLoss()
    dice_loss_fn = DiceLoss()
    optimizer = build_segnet_optimizer(model, config)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=3)

    best_val_miou = -1.0
    best_state = None
    patience_counter = 0
    history = []

    for epoch in range(1, config.max_epochs + 1):
        train_loss, epoch_time = train_one_epoch(model, train_loader, optimizer, bce_loss_fn, dice_loss_fn, device)
        val_metrics = evaluate_loader(model, val_loader, bce_loss_fn, dice_loss_fn, device, config.threshold)
        scheduler.step(val_metrics['mIoU'])

        current_lrs = [group['lr'] for group in optimizer.param_groups]
        row = {
            'epoch': epoch,
            'train_loss': train_loss,
            'val_loss': val_metrics['loss'],
            'val_mIoU': val_metrics['mIoU'],
            'val_Pet IoU': val_metrics['Pet IoU'],
            'val_Dice/F1': val_metrics['Dice/F1'],
            'epoch_time_sec': epoch_time,
            'encoder_lr': current_lrs[0],
            'decoder_lr': current_lrs[1],
        }
        history.append(row)
        display(pd.DataFrame([row]))

        if val_metrics['mIoU'] > best_val_miou:
            best_val_miou = val_metrics['mIoU']
            best_state = copy.deepcopy(model.state_dict())
            torch.save(
                {
                    'epoch': epoch,
                    'model_state': best_state,
                    'best_val_miou': best_val_miou,
                    'config': serialize_config(config),
                },
                checkpoint_path,
            )
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= config.early_stopping_patience:
            print(f'Early stopping at epoch {epoch}.')
            break

    if best_state is not None:
        model.load_state_dict(best_state)

    return pd.DataFrame(history)
```

### 19. HRNet-W18 Model


```python
def init_hrnet_head(module: nn.Module) -> None:
    if isinstance(module, nn.Conv2d):
        nn.init.kaiming_normal_(module.weight, mode="fan_out", nonlinearity="relu")
        if module.bias is not None:
            nn.init.zeros_(module.bias)
    elif isinstance(module, nn.BatchNorm2d):
        nn.init.ones_(module.weight)
        nn.init.zeros_(module.bias)


class HRNetW18BinarySeg(nn.Module):
    def __init__(
        self,
        pretrained_backbone: bool = True,
        projection_channels: int = 64,
        fusion_channels: int = 128,
    ) -> None:
        super().__init__()
        self.backbone = timm.create_model(
            "hrnet_w18",
            pretrained=pretrained_backbone,
            features_only=True,
        )
        feature_channels = list(self.backbone.feature_info.channels())
        self.feature_projections = nn.ModuleList(
            [
                nn.Sequential(
                    nn.Conv2d(channels, projection_channels, kernel_size=1, bias=False),
                    nn.BatchNorm2d(projection_channels),
                    nn.ReLU(inplace=True),
                )
                for channels in feature_channels
            ]
        )
        self.fuse = nn.Sequential(
            nn.Conv2d(projection_channels * len(feature_channels), fusion_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(fusion_channels),
            nn.ReLU(inplace=True),
        )
        self.classifier = nn.Conv2d(fusion_channels, 1, kernel_size=1)

        self.feature_projections.apply(init_hrnet_head)
        self.fuse.apply(init_hrnet_head)
        init_hrnet_head(self.classifier)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        input_size = x.shape[-2:]
        features = self.backbone(x)
        if isinstance(features, torch.Tensor):
            features = [features]
        if not features:
            raise RuntimeError("HRNet backbone returned no feature maps.")

        target_size = features[0].shape[-2:]
        projected_features = []
        for feature_map, projection in zip(features, self.feature_projections):
            projected = projection(feature_map)
            if projected.shape[-2:] != target_size:
                projected = F.interpolate(projected, size=target_size, mode="bilinear", align_corners=False)
            projected_features.append(projected)

        fused = torch.cat(projected_features, dim=1)
        fused = self.fuse(fused)
        logits = self.classifier(fused)
        return F.interpolate(logits, size=input_size, mode="bilinear", align_corners=False)
```

```python
def build_hrnet_optimizer(model: HRNetW18BinarySeg, config_obj: Config) -> torch.optim.Optimizer:
    backbone_params = []
    head_params = []

    for name, param in model.named_parameters():
        if name.startswith("backbone."):
            backbone_params.append(param)
        else:
            head_params.append(param)

    return torch.optim.AdamW(
        [
            {"params": backbone_params, "lr": config_obj.backbone_lr},
            {"params": head_params, "lr": config_obj.head_lr},
        ],
        weight_decay=config_obj.weight_decay,
    )

def save_hrnet_checkpoint(
    path: Path,
    model: nn.Module,
    optimizer: torch.optim.Optimizer,
    scheduler: torch.optim.lr_scheduler.ReduceLROnPlateau,
    epoch: int,
    best_val_miou: float,
    epochs_without_improvement: int,
    config_obj: Config,
    history_rows: list[dict[str, float]],
) -> None:
    torch.save(
        {
            "epoch": epoch,
            "model_state": copy.deepcopy(model.state_dict()),
            "optimizer_state": optimizer.state_dict(),
            "scheduler_state": scheduler.state_dict(),
            "best_val_miou": best_val_miou,
            "epochs_without_improvement": epochs_without_improvement,
            "config": serialize_config_for_storage(config_obj),
            "history": history_rows,
        },
        path,
    )

def resume_or_initialize_hrnet_training(
    model: HRNetW18BinarySeg,
    config_obj: Config,
    device: torch.device,
    latest_checkpoint_path: Path = HRNET_LATEST_MODEL_PATH,
    resume: bool = True,
) -> dict[str, object]:
    bce_loss_fn = nn.BCEWithLogitsLoss()
    dice_loss_fn = DiceLoss()
    optimizer = build_hrnet_optimizer(model, config_obj)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="max", factor=0.5, patience=3)
    start_epoch = 1
    best_val_miou = -1.0
    epochs_without_improvement = 0
    history_rows: list[dict[str, float]] = []
    resumed = False

    if resume and latest_checkpoint_path.is_file():
        checkpoint = torch.load(latest_checkpoint_path, map_location=device, weights_only=False)
        model.load_state_dict(checkpoint["model_state"])
        optimizer.load_state_dict(checkpoint["optimizer_state"])
        scheduler.load_state_dict(checkpoint["scheduler_state"])
        move_optimizer_state_to_device(optimizer, device)
        start_epoch = int(checkpoint["epoch"]) + 1
        best_val_miou = float(checkpoint.get("best_val_miou", -1.0))
        epochs_without_improvement = int(checkpoint.get("epochs_without_improvement", 0))
        history_rows = list(checkpoint.get("history", []))
        resumed = True

    return {
        "optimizer": optimizer,
        "scheduler": scheduler,
        "bce_loss_fn": bce_loss_fn,
        "dice_loss_fn": dice_loss_fn,
        "start_epoch": start_epoch,
        "best_val_miou": best_val_miou,
        "epochs_without_improvement": epochs_without_improvement,
        "history_rows": history_rows,
        "resumed": resumed,
    }
```

```python
def fit_hrnet_model(
    model: HRNetW18BinarySeg,
    train_loader: DataLoader,
    val_loader: DataLoader,
    config_obj: Config,
    device: torch.device,
    resume: bool = True,
    latest_checkpoint_path: Path = HRNET_LATEST_MODEL_PATH,
    best_checkpoint_path: Path = HRNET_BEST_MODEL_PATH,
    history_csv_path: Path = HRNET_HISTORY_CSV,
) -> pd.DataFrame:
    training_state = resume_or_initialize_hrnet_training(
        model=model,
        config_obj=config_obj,
        device=device,
        latest_checkpoint_path=latest_checkpoint_path,
        resume=resume,
    )
    optimizer = training_state["optimizer"]
    scheduler = training_state["scheduler"]
    bce_loss_fn = training_state["bce_loss_fn"]
    dice_loss_fn = training_state["dice_loss_fn"]
    start_epoch = int(training_state["start_epoch"])
    best_val_miou = float(training_state["best_val_miou"])
    epochs_without_improvement = int(training_state["epochs_without_improvement"])
    history_rows = list(training_state["history_rows"])

    if training_state["resumed"]:
        print(f"Resuming HRNet-W18 from epoch {start_epoch}.")
    else:
        print("Starting a fresh HRNet-W18 run.")

    if start_epoch > config_obj.max_epochs:
        print(f"HRNet-W18 is already trained through epoch {start_epoch - 1}.")
        return save_history_csv(history_rows, history_csv_path)

    for epoch in range(start_epoch, config_obj.max_epochs + 1):
        train_loss, epoch_time = train_one_epoch(model, train_loader, optimizer, bce_loss_fn, dice_loss_fn, device)
        val_metrics = evaluate_loader(model, val_loader, bce_loss_fn, dice_loss_fn, device, config_obj.threshold)
        scheduler.step(val_metrics["mIoU"])
        current_lrs = [group["lr"] for group in optimizer.param_groups]
        row = {
            "epoch": epoch,
            "train_loss": train_loss,
            "val_loss": val_metrics["loss"],
            "val_mIoU": val_metrics["mIoU"],
            "val_Pet IoU": val_metrics["Pet IoU"],
            "val_Dice/F1": val_metrics["Dice/F1"],
            "epoch_time_sec": epoch_time,
            "backbone_lr": current_lrs[0],
            "head_lr": current_lrs[1],
        }
        history_rows.append(row)
        save_history_csv(history_rows, history_csv_path)

        improved = val_metrics["mIoU"] > best_val_miou
        if improved:
            best_val_miou = float(val_metrics["mIoU"])
            epochs_without_improvement = 0
            save_hrnet_checkpoint(best_checkpoint_path, model, optimizer, scheduler, epoch, best_val_miou, epochs_without_improvement, config_obj, history_rows)
        else:
            epochs_without_improvement += 1

        save_hrnet_checkpoint(latest_checkpoint_path, model, optimizer, scheduler, epoch, best_val_miou, epochs_without_improvement, config_obj, history_rows)
        print(f"Epoch {epoch:02d}/{config_obj.max_epochs} | train_loss={train_loss:.6f} | val_loss={val_metrics['loss']:.6f} | val_mIoU={val_metrics['mIoU']:.6f}")

        if epochs_without_improvement >= config_obj.early_stopping_patience:
            print(f"Early stopping at epoch {epoch}.")
            break

    history_df_local = save_history_csv(history_rows, history_csv_path)
    if best_checkpoint_path.is_file():
        best_checkpoint = torch.load(best_checkpoint_path, map_location=device, weights_only=False)
        model.load_state_dict(best_checkpoint["model_state"])
    return history_df_local

def sort_model_results(model_results_df: pd.DataFrame) -> pd.DataFrame:
    if model_results_df.empty:
        return model_results_df
    sorted_df = model_results_df.copy()
    sorted_df["__order__"] = sorted_df["Model"].map({name: idx for idx, name in enumerate(MODEL_RESULTS_ORDER)})
    sorted_df["__order__"] = sorted_df["__order__"].fillna(len(MODEL_RESULTS_ORDER)).astype(int)
    sorted_df = sorted_df.sort_values(["__order__", "Model"]).drop(columns="__order__")
    return sorted_df.reset_index(drop=True)

def ensure_model_results_store(model_results_csv: Path = MODEL_RESULTS_CSV) -> pd.DataFrame:
    if model_results_csv.is_file():
        existing_df = pd.read_csv(model_results_csv)
    else:
        existing_df = pd.DataFrame(columns=MODEL_RESULTS_COLUMNS)
    combined_df = pd.concat([existing_df, BASELINE_MODEL_ROWS], ignore_index=True)
    combined_df = combined_df.drop_duplicates(subset=["Model"], keep="last")
    combined_df = combined_df[MODEL_RESULTS_COLUMNS]
    combined_df = sort_model_results(combined_df)
    combined_df.to_csv(model_results_csv, index=False)
    return combined_df

def upsert_model_results(new_rows_df: pd.DataFrame, model_results_csv: Path = MODEL_RESULTS_CSV) -> pd.DataFrame:
    combined_df = ensure_model_results_store(model_results_csv)
    combined_df = pd.concat([combined_df, new_rows_df], ignore_index=True)
    combined_df = combined_df.drop_duplicates(subset=["Model"], keep="last")
    combined_df = combined_df[MODEL_RESULTS_COLUMNS]
    combined_df = sort_model_results(combined_df)
    combined_df.to_csv(model_results_csv, index=False)
    return combined_df
```

### 23. HRNet Final Test Evaluation and Efficiency Metrics


```python
def measure_inference_time_hrnet(
    model: nn.Module,
    loader: DataLoader,
    device: torch.device,
    warmup: int = 10,
    timed: int = 100,
) -> float:
    model.eval()
    timings = []
    seen = 0

    with torch.no_grad():
        for idx, (batch_images, _) in enumerate(loader):
            batch_images = batch_images.to(device)
            start = perf_counter()
            _ = model(batch_images)
            elapsed = perf_counter() - start
            if idx >= warmup:
                timings.append(elapsed / batch_images.size(0))
                seen += batch_images.size(0)
            if seen >= timed:
                break
    return float(np.mean(timings)) if timings else float("nan")

if not HRNET_BEST_MODEL_PATH.is_file():
    raise FileNotFoundError(f"Missing HRNet best checkpoint: {HRNET_BEST_MODEL_PATH}")

hrnet_checkpoint = torch.load(HRNET_BEST_MODEL_PATH, map_location=DEVICE, weights_only=False)
hrnet_model = HRNetW18BinarySeg(pretrained_backbone=hrnet_config.use_pretrained_backbone).to(DEVICE)
hrnet_model.load_state_dict(hrnet_checkpoint["model_state"])

bce_loss_fn = nn.BCEWithLogitsLoss()
dice_loss_fn = DiceLoss()
hrnet_test_metrics = evaluate_loader(hrnet_model, test_loader, bce_loss_fn, dice_loss_fn, DEVICE, hrnet_config.threshold)

if HRNET_HISTORY_CSV.is_file():
    hrnet_history_metrics_df = pd.read_csv(HRNET_HISTORY_CSV)
    hrnet_avg_epoch_time = float(hrnet_history_metrics_df["epoch_time_sec"].mean())
else:
    hrnet_avg_epoch_time = float("nan")

hrnet_inference_time = measure_inference_time_hrnet(hrnet_model, test_loader, DEVICE)
hrnet_total_params = sum(param.numel() for param in hrnet_model.parameters())
hrnet_results_df = pd.DataFrame(
    [
        {
            "Model": "HRNet-W18",
            "Backbone Init": "pretrained",
            "Test mIoU": hrnet_test_metrics["mIoU"],
            "Pet IoU": hrnet_test_metrics["Pet IoU"],
            "Dice/F1": hrnet_test_metrics["Dice/F1"],
            "Pixel Accuracy": hrnet_test_metrics["Pixel Accuracy"],
            "Foreground Precision": hrnet_test_metrics["Foreground Precision"],
            "Foreground Recall": hrnet_test_metrics["Foreground Recall"],
            "Parameter Count": hrnet_total_params,
            "Inference Time / Image (sec)": hrnet_inference_time,
            "Training Time / Epoch (sec)": hrnet_avg_epoch_time,
        }
    ],
    columns=MODEL_RESULTS_COLUMNS,
)
comparison_df_3_models = upsert_model_results(hrnet_results_df)
display(hrnet_results_df)
```

### 24. HRNet Qualitative Predictions


```python
def collect_predictions_for_display(
    model: nn.Module,
    loader: DataLoader,
    device: torch.device,
    threshold: float,
    limit: int = 4,
) -> list[dict[str, np.ndarray]]:
    model.eval()
    collected = []
    with torch.no_grad():
        for batch_images, batch_masks in loader:
            batch_images = batch_images.to(device)
            logits = model(batch_images)
            preds = (torch.sigmoid(logits) >= threshold).float().cpu()
            for image_tensor, mask_tensor, pred_tensor in zip(batch_images.cpu(), batch_masks, preds):
                collected.append({
                    "image": denormalize_image(image_tensor),
                    "mask": mask_tensor.squeeze(0).numpy(),
                    "pred": pred_tensor.squeeze(0).numpy(),
                })
                if len(collected) >= limit:
                    return collected
    return collected

def show_predictions_grid(prediction_items: list[dict[str, np.ndarray]]) -> None:
    rows = len(prediction_items)
    fig, axes = plt.subplots(rows, 4, figsize=(16, 4 * rows))
    axes = np.atleast_2d(axes)
    for row_axes, item in zip(axes, prediction_items):
        image = item["image"]
        mask = item["mask"]
        pred = item["pred"]
        overlay = 0.65 * image + 0.35 * np.stack([pred, np.zeros_like(pred), np.zeros_like(pred)], axis=-1)
        row_axes[0].imshow(image)
        row_axes[0].set_title("Input Image")
        row_axes[1].imshow(mask, cmap="gray")
        row_axes[1].set_title("Ground Truth")
        row_axes[2].imshow(pred, cmap="gray")
        row_axes[2].set_title("Prediction")
        row_axes[3].imshow(np.clip(overlay, 0.0, 1.0))
        row_axes[3].set_title("Prediction Overlay")
        for axis in row_axes:
            axis.axis("off")
    plt.tight_layout()
    plt.show()

if not HRNET_BEST_MODEL_PATH.is_file():
    raise FileNotFoundError(f"Missing HRNet best checkpoint: {HRNET_BEST_MODEL_PATH}")

hrnet_checkpoint = torch.load(HRNET_BEST_MODEL_PATH, map_location=DEVICE, weights_only=False)
hrnet_model = HRNetW18BinarySeg(pretrained_backbone=hrnet_config.use_pretrained_backbone).to(DEVICE)
hrnet_model.load_state_dict(hrnet_checkpoint["model_state"])
hrnet_prediction_items = collect_predictions_for_display(hrnet_model, test_loader, DEVICE, hrnet_config.threshold, limit=4)
show_predictions_grid(hrnet_prediction_items)
```
